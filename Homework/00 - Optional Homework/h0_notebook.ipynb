```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("husl")
```

```python
def generate_pie_plot(data, title, labels):
    plt.title(title)
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=140)
```

```python
data_path = 'data/interactive_data.csv'

with open(data_path, 'r') as f:
    data = pd.read_csv(f)

data = data.dropna()

xdata = data['Intent'].value_counts().values
labels = data['Intent'].value_counts().index.values
generate_pie_plot(xdata, f"Intent of Gun Deaths over {data.size} responses", labels)
```
